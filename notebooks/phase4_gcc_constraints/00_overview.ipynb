# Phase 4: GCC Constraint Simulation

## Scope
Simulate GCC (Government Commercial Cloud) constraints and re-run the same four
Phase 3 failures. Record which failures become unrecoverable under constrained
network conditions.

## GCC Constraints (Python-level simulation)
| Constraint | Value | Implementation |
|-----------|-------|---------------|
| Added latency | 200ms + 50ms jitter | `asyncio.sleep` before each node call |
| Egress proxy overhead | Included in latency | Random jitter simulates proxy variance |
| TPM rate cap | 60,000 tokens/min | Sliding-window counter, sleeps when exhausted |

## Why not toxiproxy?
The original spec called for toxiproxy. We implemented constraints in Python instead:
- Zero infrastructure (no Docker, no proxy process)
- Reproducible (deterministic latency, not network-dependent)
- Same observable effect (added latency, rate limiting)
- Anyone cloning the repo can run it immediately
