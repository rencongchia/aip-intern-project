# Phase 3: Fault Injection

## Scope
Inject four failures into the LangGraph+CrewAI mesh and score each on recovery time,
output quality, and whether recovery was automatic or manual.

## Fault Types
| Fault | Injector | Recovery Expected |
|-------|----------|-------------------|
| `timeout` | `asyncio.wait_for` with pre-sleep | manual (AgentTimeoutError propagates) |
| `malformed_json` | Proxy crew raises JSONDecodeError | automatic (crew_node catches it) |
| `checkpoint_loss` | Crew runs, then CheckpointLostError raised | manual (state update lost) |
| `context_overflow` | Token count pre-check exceeds limit | manual (before crew runs) |

## Key Files
- Injectors: `src/aip_intern/failures/injectors.py`
- Scoring: `src/aip_intern/failures/scoring.py`
- Runner: `src/aip_intern/phase3/runner.py`
- Config: `config/phase3.yaml`
- CLI: `scripts/run_phase3.py`

## How to run
```bash
# Dry-run (1 per fault type)
python scripts/run_phase3.py --config config/phase3.yaml --dry-run

# Full sweep (5 per fault type = 20 total)
python scripts/run_phase3.py --config config/phase3.yaml

# Single fault type
python scripts/run_phase3.py --config config/phase3.yaml --fault timeout
```
